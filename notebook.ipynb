```r
data <- read.csv('product_sales.csv')
data
```

```r
# Check data types of columns
str(data)
```

```r
# Ensure correct column types
data$week <- as.numeric(data$week)  # Numeric
data$sales_method <- as.factor(data$sales_method)  # Factor
data$customer_id <- as.character(data$customer_id)  # Character
data$nb_sold <- as.numeric(data$nb_sold)  # Numeric
data$revenue <- as.numeric(data$revenue)  # Numeric
data$years_as_customer <- as.numeric(data$years_as_customer)  # Numeric
data$nb_site_visits <- as.numeric(data$nb_site_visits)  # Numeric
data$state <- as.factor(data$state)  # Factor
```

```r
# Check for missing values
colSums(is.na(data))

# # Summary of the dataset
# summary(data)
```

```r
# Check for missing values in the 'revenue' column
sum(is.na(data$revenue))
```

```r
data$revenue[is.na(data$revenue)] <- mean(data$revenue, na.rm = TRUE)
#Replaced the missing values with the overall mean of the data in revenue
```

```r
sum(is.na(data$revenue))
```

```r
data$sales_method <- factor(data$sales_method)
levels(data$sales_method)
```

```r
# Clean the sales_method column
data$sales_method <- factor(data$sales_method) # Ensure it's a factor

# Replace categories
data$sales_method <- as.character(data$sales_method) # Convert factor to character for replacement
data$sales_method[data$sales_method == "em + call"] <- "Email + Call"
data$sales_method[data$sales_method == "call"] <- "Call"
data$sales_method[data$sales_method == "email"] <- "Email"

# Convert back to a factor with the cleaned categories
data$sales_method <- factor(data$sales_method)

# Check the levels
levels(data$sales_method)
```

```r
# Check for duplicate values in customer_id
duplicates <- duplicated(data$customer_id)

# If there are duplicates, the sum of TRUE values in `duplicates` will be greater than 0
if (sum(duplicates) == 0) {
  print("No duplicate values found in customer_id.")
} else {
  print(paste("Number of duplicate values in customer_id:", sum(duplicates)))
}
```

```r
# Columns to check for non-numeric values
columns_to_check <- c("week", "nb_sold", "revenue", "years_as_customer", "nb_site_visits")

# Loop through the columns and check for non-numeric values
for (col in columns_to_check) {
  # Check if all values in the column are numeric
  if (all(is.numeric(data[[col]]))) {
    print(paste("Column", col, "contains only numeric values."))
  } else {
    print(paste("Column", col, "contains non-numeric values."))
    # Print the non-numeric values, if any
    non_numeric_values <- data[[col]][!is.numeric(data[[col]])]
    print(non_numeric_values)
  }
}
```

```r
# Load necessary libraries
library(ggplot2)
library(dplyr)
```

```r
# Count of customers for each sales method
customer_count <- table(data$sales_method)
print(customer_count)

# Bar plot for customer count by sales method
ggplot(data, aes(x = sales_method)) +
  geom_bar(fill = "blue") +
  labs(title = "Number of Customers by Sales Method", x = "Sales Method", y = "Count")
```

```r
# Summary statistics for revenue overall
print(summary(data$revenue))

# Boxplot for revenue by sales method
ggplot(data, aes(x = sales_method, y = revenue)) +
  geom_boxplot(fill = "lightblue") +
  labs(title = "Revenue Spread by Sales Method", x = "Sales Method", y = "Revenue")
  
# Histogram of revenue overall
ggplot(data, aes(x = revenue)) +
  geom_histogram(binwidth = 50, fill = "blue", color = "black", alpha = 0.7) +
  labs(title = "Histogram of Revenue", x = "Revenue", y = "Frequency")
```

```r
# Aggregate revenue by week and sales method
revenue_trend <- data %>%
  group_by(week, sales_method) %>%
  summarise(total_revenue = sum(revenue, na.rm = TRUE))

# Line plot for revenue over time by sales method
ggplot(revenue_trend, aes(x = week, y = total_revenue, color = sales_method)) +
  geom_line(size = 1) +
  labs(title = "Revenue Trend Over Time by Sales Method", x = "Week", y = "Total Revenue")
```

```r
# Calculate average revenue by sales method
avg_revenue <- data %>%
  group_by(sales_method) %>%
  summarise(average_revenue = mean(revenue, na.rm = TRUE),
            std_dev_revenue = sd(revenue, na.rm = TRUE),
            customer_count = n())

print(avg_revenue)

# Bar plot for average revenue by sales method
ggplot(avg_revenue, aes(x = sales_method, y = average_revenue, fill = sales_method)) +
  geom_bar(stat = "identity", color = "black", alpha = 0.7) +
  labs(title = "Average Revenue by Sales Method", x = "Sales Method", y = "Average Revenue")
```

```r
# Load required libraries
library(ggplot2)
library(dplyr)

# Aggregate data: calculate the mean number of site visits for each sales method
site_visits_by_method <- data %>%
  group_by(sales_method) %>%
  summarise(avg_site_visits = mean(nb_site_visits, na.rm = TRUE))
print(site_visits_by_method)
# Create a bar chart
ggplot(site_visits_by_method, aes(x = sales_method, y = avg_site_visits, fill = sales_method)) +
  geom_bar(stat = "identity") +
  theme_minimal() +
  labs(
    title = "Average Number of Site Visits by Sales Method",
    x = "Sales Method",
    y = "Average Number of Site Visits"
  ) +
  scale_fill_brewer(palette = "Set2")
```

```r
# Rename dataframe
sales_data <- data

# Calculate Revenue Per Customer (RPC)
rpc_summary <- sales_data %>%
  group_by(sales_method) %>%
  summarise(
    total_revenue = sum(revenue, na.rm = TRUE),
    total_customers = n(),
    revenue_per_customer = total_revenue / total_customers
  )

# Print the RPC summary
print(rpc_summary)



# Bar chart for Revenue Per Customer (RPC) by sales_method
ggplot(rpc_summary, aes(x = sales_method, y = revenue_per_customer, fill = sales_method)) +
  geom_bar(stat = "identity") +
  theme_minimal() +
  labs(
    title = "Revenue Per Customer by Sales Method",
    x = "Sales Method",
    y = "Revenue Per Customer"
  ) +
  scale_fill_brewer(palette = "Set2")
```

```r
# Revenue by Sales Method
revenue_by_method <- aggregate(revenue ~ sales_method, data = data, sum)

# Average Revenue per Sales Method
avg_revenue_method <- aggregate(revenue ~ sales_method, data = data, mean)

# Products Sold by Sales Method
products_by_method <- aggregate(nb_sold ~ sales_method, data = data, sum)

# Combine Metrics
sales_channel_metrics <- merge(revenue_by_method, avg_revenue_method, by = "sales_method")
sales_channel_metrics <- merge(sales_channel_metrics, products_by_method, by = "sales_method")
colnames(sales_channel_metrics) <- c("Sales_Method", "Total_Revenue", "Average_Revenue", "Total_Products_Sold")
print(sales_channel_metrics)
```

```r
# Bar Plot for Revenue by State
# Revenue by State
# Aggregate revenue by state
revenue_by_state <- aggregate(revenue ~ state, data = data, sum)

# Order the data frame by revenue in descending order
revenue_by_state <- revenue_by_state[order(-revenue_by_state$revenue), ]

# Print the ordered data frame
print(revenue_by_state)

ggplot(revenue_by_state, aes(x = reorder(state, revenue), y = revenue, fill = state)) +
  geom_bar(stat = "identity") +
  coord_flip() +
  labs(title = "Revenue by State", x = "State", y = "Revenue") +
  theme_minimal()
```
